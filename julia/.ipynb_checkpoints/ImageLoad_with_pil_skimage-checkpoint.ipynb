```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
def plot_sample(x, y, axis):
    img = x.reshape(96,96)
    axis.imshow(img)
    axis.axis('off')

def draw_multi_plot(X, y, row_num=4, col_num=4, fig_size=(15,10)):
    fig, axes = plt.subplots(nrows=row_num, ncols=col_num, figsize=fig_size)
    i=0
    for row in axes:
        for each_ax in row:
            plot_sample(X[i], y[i], each_ax)
            i += 1
    plt.tight_layout()
```
